```python
from statistics import mean
```

```python
#function to view the breakdown of ethnicity by grade
def stu_demo(grade):
    avg_asian=round(mean(student_df['% Asian']), 2)
    avg_black=round(mean(student_df['% Black']), 2)
    avg_hisp=round(mean(student_df['% Hispanic']),2)
    avg_white=round(mean(student_df['% White']),2)
    
    avg_demo={'Asian': avg_asian, 
              'Black': avg_black,
              'Hispanic': avg_hisp,
              'White': avg_white}
    return avg_demo
```

```python
#group data by year and sum
group_df=student_df.groupby(by=['Year']).sum()
group_df.head()
```

```python
#group data by school and sum
group_school=student_df.groupby(by=['School Name']).sum()
group_school.head()
```

```python
group_df.reset_index(inplace=True)
```

```python
type(group_df.Year)
```

```python
plt.rcParams["figure.figsize"] = [10, 8]
```

### Ethnicity

```python
graph1=group_df.plot(x="Year", y=["# Asian", "# Black", "# Hispanic", "# White"], kind="bar")
for p in graph1.patches:
    height = p.get_height()
    graph1.text(p.get_x()+p.get_width()/2., height - 100000,height ,ha="center", 
                rotation=90, color='white', fontsize=13)
```

### Gender

```python
graph2=group_df.plot(x="Year", y=["# Male", "# Female"], kind="bar")
for p in graph2.patches:
    height = p.get_height()
    graph2.text(p.get_x()+p.get_width()/2., height - 100000,height ,ha="center", 
                rotation=90, color='white', fontsize=13)
```
